## Step 1: Data Loading
In this step, we import the `pandas` library and load the video game sales dataset (2024 version) from a CSV file.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgchartz-2024.csv')

df.head()
```

```python
nintendo_df = df[df['publisher'] == 'Nintendo'].copy()

cols_to_keep = ['title', 'console', 'genre', 'total_sales', 'na_sales', 'jp_sales', 'pal_sales']
nintendo_df = nintendo_df[cols_to_keep].dropna(subset=['total_sales'])

nintendo_df = nintendo_df.sort_values(by='total_sales', ascending=False)

print(f"Successfully filtered {len(nintendo_df)} Nintendo games!")
nintendo_df.head(10)
```

## Step 3: Visualisation and Insights
To communicate analytical value to stakeholders, I created a bar chart showing the top 10 best-selling Nintendo games globally. 
This helps identify which franchises drive the most revenue.

```python
top_10_games = nintendo_df.head(10)

plt.figure(figsize=(12, 6))
plt.barh(top_10_games['title'], top_10_games['total_sales'], color='red')
plt.xlabel('Total Sales (Millions)')
plt.title('Top 10 Best-Selling Nintendo Games (Business Insight)')
plt.gca().invert_yaxis()  # 让排名第一的在最上面
plt.show()
```
